```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('Profit ~ Spend+Administration+Marketing_Spend+State_California+State_Florida+State_NewYork',data=df2).fit()
```

```python
model.params
```

```python
model.rsquared
```

```python
model.rsquared_adj
```

```python
a=model.resid
```

```python
model.predict(df2.iloc[0:3,0:])
```
