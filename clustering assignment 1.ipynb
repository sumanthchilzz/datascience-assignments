# k means

```python
df1
```

```python
from sklearn.cluster import KMeans
```

```python
#distortion
distortions = []
K = range(1,10)
for i in K:
    kmeanModel = KMeans(n_clusters=i)
    kmeanModel.fit(df1)
    distortions.append(kmeanModel.inertia_)
```

```python
#ploting the elbow curve and determing the optimal k-value
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

### implementing PCA technique

```python
from sklearn.decomposition import PCA
```

```python
#creating and fitting the model to a dataset
pca = PCA(n_components=4)
pca_values = pca.fit_transform(norm1)
```

```python
pca_values
```

```python
df7=pd.DataFrame(pca_values,columns=['pc1','pc2','pc3','pc4'])
df7
```

```python
#variance ratio
var = pca.explained_variance_ratio_
```

```python
var
```

```python
#Return the cumulative sum of the elements
var1= np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="red",marker='o',linestyle='--')
```
